# İlk Git (Git) Depomuzu Oluşturma ve İlk Commit (Commit)

Artık Git’in ne olduğunu, ne işe yaradığını ve nasıl kurulduğunu öğrendiğimize göre, ilk Git **repository**’mizi (depo) oluşturabiliriz.
**Repository**, üzerinde Git ile çalıştığımız dosya sistemindeki dizindir. Yani öncelikle bu çalışma alanını oluşturmamız gerekir.

Hatırlarsanız, Git’te iki tür repository vardır:

* **Local repository** (yerel depo)
* **Remote repository** (uzak depo)

İlk aşamada her şey **lokal** olarak yapılır, ardından **remote** ortama aktarılır. Bu yüzden öncelikle **local repository**’yi oluşturacağız. Bu yazıda, local repository oluşturmayı, dosya eklemeyi, depo durumunu kontrol etmeyi ve ilk commit’i yapmayı öğreneceksiniz.

---

## 1. Git init (git init)

Git, dosyalarını ve geçmişini proje klasöründe saklar.
Yeni bir repository oluşturmak için terminali açın, proje klasörünüze gidin ve şu komutu çalıştırın:

```bash
git init
```

Bu komut, o klasörde Git’i başlatır ve `.git` adında gizli bir klasör oluşturur. Bu klasör, repository’nin geçmişini ve ayarlarını saklar.

Komut çıktısı şu şekilde olur:

```
Initialized empty Git repository in /klasör_yolu/repository_adi/.git/
```

Bu, repository’nin başarıyla oluşturulduğu anlamına gelir, ancak şu an boş. O zaman dosya ekleyelim.

---

## 2. Git add (git add)

Repository içine yeni bir dosya oluşturabilir veya daha önce yazılmış bir programı kaydedebilirsiniz.
İki yöntemle dosya eklenebilir:

* Tek tek dosya ekleme:

```bash
git add my_file.txt
```

* Tüm dosyaları ekleme:

```bash
git add -A
```

Bu şekilde dosyalar **staging area**’ya (hazırlık alanı) eklenmiş olur. Ama işimiz bitmedi.

---

## 3. Repository durumunu kontrol etme – Git status (git status)

Yaptığımız işlemlerin doğru olduğunu görmek ve değişiklikleri takip etmek için `git status` komutunu kullanırız:

```bash
git status
```

Örnek çıktı:

```
On branch master
Initial commit
Changes to be committed:
  (use "git rm --cached ..." to unstage)
        new file: my_file.txt
```

Bu çıktı bize, repository’de yeni bir dosya olduğunu gösterir.
`git status` komutunu, çalışma sürecinizin her aşamasında kullanabilirsiniz.

---

## 4. Git commit (git commit)

**Commit**, repository’nin belirli bir zamandaki durumunu temsil eder.
Yapılan değişiklikleri kayıt altına alır ve sonradan takip etmeyi kolaylaştırır.

Commit işlemi yapmak için staging area’da en az bir değişiklik olması gerekir (önceki adımda dosya eklediğimiz gibi). Ardından şu komutu çalıştırırız:

```bash
git commit -m "First commit."
```

Buradaki `-m`, **message** (mesaj) anlamına gelir ve commit’e yorum eklemenizi sağlar.
Anlamlı commit mesajları yazmak iyi bir alışkanlıktır; hem sizin hem de ekibinizin ileride yaptığı değişiklikleri anlamasını kolaylaştırır.

---

## Sonuç

Bu yazıda şunları öğrendik:

* Yeni repository oluşturmak için:

```bash
git init
```

* Dosya eklemek için:

```bash
git add <dosya_adi.uzanti>
```

veya

```bash
git add -A
```

* Değişiklikleri commit etmek için:

```bash
git commit -m "Mesaj"
```

* Depo durumunu kontrol etmek için:

```bash
git status
```
