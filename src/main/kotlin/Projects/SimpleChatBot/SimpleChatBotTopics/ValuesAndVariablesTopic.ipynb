

---

### 🔹 **Değişken (Variable) Nedir?**

* Değişken, bir **değeri saklamak** için kullanılan isimlendirilmiş bir yapıdır.
* Bu değer bir **sayı**, **metin (string)**, **karakter** vb. olabilir.
* Değişkenler sayesinde veriye **isimle erişebiliriz**.

---

### 🔑 **Değişken Tanımlama (Declaration)**

#### ✅ `val`

* **Sabit (değiştirilemez)** bir değişken tanımlar.
* Sadece bir kez değer atanabilir.

```kotlin
val language = "Kotlin"
```

#### ✅ `var`

* **Değiştirilebilir (mutable)** bir değişken tanımlar.
* Değeri daha sonra güncellenebilir.

```kotlin
var dayOfWeek = "Monday"
dayOfWeek = "Tuesday"
```

#### ✅ `const val`

* Derleme zamanında bilinen sabitler için kullanılır (ileride daha detaylı öğrenilecektir).

---

### 📛 **Değişken İsimleri**

* Sayı ile **başlayamaz**.
* **Anlamlı ve okunabilir** olmalı.
* **Büyük-küçük harf duyarlıdır** (`language` ≠ `Language`).

---

### ⚙️ **Değer Atama**

* Atama işlemi `=` operatörü ile yapılır.

```kotlin
val cost = 3
val costOfCoffee = cost
```

---

### 💡 **Farklı Veri Türlerini Saklamak**

Değişkenler farklı türlerde veri saklayabilir:

```kotlin
val ten = 10           // sayı
val greeting = "Hello" // string
val firstLetter = 'A'  // karakter
```

---

### ❗ **`var` ile Tür Değişikliği Yapılamaz**

Bir `var` değişkeni tanımlandıktan sonra, **aynı türden** değerlerle güncellenebilir.

```kotlin
var number = 10
number = 11       // ✅
number = "twelve" // ❌ Hata: Tür uyumsuzluğu
```

---

### 🧾 **Sonuç**

* **`val`**: Değeri değişmeyen sabitler için (tercih edilmesi önerilir).
* **`var`**: Değeri zamanla değişecek olanlar için.
* **Okunabilir, güvenli ve sade kod yazmak için mümkün olduğunca `val` kullanın.**
