### ✅ **Boolean Veri Tipi Nedir?**

* Sadece iki değeri olabilir: `true` (doğru) veya `false` (yanlış).
* Örneğin:

  ```kotlin
  val isOpen = true
  val isClosed = false
  ```
* Kotlin’de `0` veya `1` gibi sayılar Boolean yerine geçmez.

---

### ✅ **Boolean Değer Okuma**

* `readln().toBoolean()` yöntemi ile kullanıcıdan `true` veya `false` değerleri okunabilir.

  * Büyük/küçük harf duyarsızdır. Örneğin: `"True"` → `true`
* `toBooleanStrict()` → sadece `"true"` ve `"false"` kabul eder, aksi halde hata verir.
* `toBooleanStrictOrNull()` → sadece `"true"`/`"false"` döner, diğer girişlerde `null` döner.

✅ **Mantıksal Operatörler**

1. **NOT (`!`)**: Mantıksal değeri tersine çevirir.

   ```kotlin
   val t = !false // true
   ```

2. **AND (`&&`)**: Her iki değer de `true` ise sonuç `true`.

   ```kotlin
   true && true → true
   ```

3. **OR (`||`)**: Değerlerden biri `true` ise sonuç `true`.

   ```kotlin
   false || true → true
   ```

4. **XOR (`xor`)**: Değerler birbirinden farklıysa `true`, aynıysa `false`.

   ```kotlin
   true xor false → true
   ```

---

### ✅ **Operatör Önceliği (Sıralama)**

Kotlin'de operatörlerin çalıştırılma sırası:

1. `!` (NOT)
2. `xor`
3. `&&` (AND)
4. `||` (OR)

Parantezlerle bu sıralama değiştirilebilir.

---

### ✅ **Uygulama Örneği**

```kotlin
val cold = false
val dry = true
val summer = false

val hiking = dry && (!cold || summer) // true → yürüyüşe çıkılabilir
```

### 🎯 **Sonuç**

* `Boolean` tipi programlamada çok kullanılır (karar verme, durum kontrolü).
* Kotlin’de `!`, `&&`, `||`, `xor` gibi mantıksal operatörleri doğru anlamak çok önemlidir.
* Operatör önceliklerine dikkat etmek gerekir, aksi takdirde beklenmeyen sonuçlar çıkabilir.
