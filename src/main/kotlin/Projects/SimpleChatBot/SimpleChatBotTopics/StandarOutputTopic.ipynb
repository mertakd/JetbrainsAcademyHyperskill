

---

### 🖨️ **Standart Çıktı (Standard Output) Nedir?**

* Programların ekrana bilgi yazdırdığı temel işlemdir.
* Genellikle ekran kullanılır ama çıktı bir dosyaya da yönlendirilebilir.
* Hyperskill’de genellikle metin veya sayılar yazdırılır.

---

### 🔹 **println() ve print() Fonksiyonları**

#### ✅ `println()`

* Ekrana metin yazar ve **satır atlar**.

```kotlin
println("Hello")
```

**Çıktı:**

```
Hello
```

#### ✅ `print()`

* Ekrana metin yazar ama **satır atlamaz**, aynı satıra devam eder.

```kotlin
print("Hello ")
print("World")
```

**Çıktı:**

```
Hello World
```

#### 🆕 Boş satır yazdırmak için:

```kotlin
println() // sadece satır atlar
```

---

### 🔢 **Sayılar ve Karakterler Yazdırmak**

* `println` ve `print`, **sayıları** ve **karakterleri** de yazdırabilir.

```kotlin
print(108)
print('A')
println("Z")
```

**Çıktı:**

```
108AZ
```

---

### 💲 **\$ Operatörü (String Template)**

#### 🔸 Değişken değeri yazdırmak:

```kotlin
val name = "Alice"
println("Hello, $name!") // Hello, Alice!
```

#### 🔸 İfade (expression) yazdırmak:

```kotlin
val a = 5
val b = 10
println("Sum is ${a + b}") // Sum is 15
```

#### 🔸 Dolar sembolü (\$) yazdırmak:

```kotlin
println("Price is \$20")       // Price is $20
println("Price is $$a")        // Price is $20
println("Price is \$$a")       // Price is $20
```

---

### 🧾 **Sonuç:**

* `println()` satır sonu ile yazdırır, `print()` aynı satıra devam eder.
* Hem metin, hem sayılar, hem karakterler yazdırılabilir.
* `$` operatörüyle değişken veya ifadeler string içine kolayca gömülebilir.
* `$` sembolünü yazmak için `\$` kullanılır.
