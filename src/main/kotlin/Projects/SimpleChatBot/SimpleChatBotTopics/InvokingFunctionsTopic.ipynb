

---

### 📌 **Fonksiyon Nedir?**

Fonksiyon, belirli bir görevi gerçekleştiren, gerektiğinde dışarıdan veri alıp sonuç döndürebilen bir alt programdır. Program içerisinde ismiyle çağrılarak çalıştırılır.

---

### 📥 **Fonksiyon Nasıl Çağrılır?**

Fonksiyonlar, adlarının sonuna parantez eklenerek çağrılır. Parantez içine **gerekirse** veri (argüman) verilir.

#### ✅ Örnek:


```kotlin
val text = "Hello"
println(text)  // Konsola "Hello" yazdırır
```

#### ✅ Argüman olmadan:

```kotlin
println() // Sadece boş bir satır yazdırır
```

#### ✅ Genel çağırma yapısı:

```kotlin
function1()             // Argümansız
function2(arg1)         // Tek argümanlı
function3(arg1, arg2)   // İki argümanlı
```

---

### 📤 **Sonuç Döndüren Fonksiyonlar**

Bazı fonksiyonlar sadece işlem yapmaz, bir sonuç da döner. Bu sonucu bir değişkene atayabiliriz.

#### ✅ Örnek:

```kotlin
val number = -10
val nonNegNumber = Math.abs(number)  // Sonuç: 10
```

`Math.abs` fonksiyonu negatif sayıyı pozitif hale getirir.

---

### ⚠️ **Bazı Fonksiyonlar “Sonuçsuz” Gibi Görünebilir**

Ama aslında her fonksiyon bir sonuç döner. Eğer sonuç döndürmüyor gibi görünüyorsa, `Unit` döndürüyordur. Bu, Kotlin'de "hiçbir şey" anlamına gelir (Java/C'deki `void` gibi).

#### ✅ Örnek:

```kotlin
val result = println("text")
println(result) // Çıktı: kotlin.Unit
```

---

### 📌 **Sonuç (Özet)**

* Fonksiyonlar, belirli işleri yapmak için kullanılır.
* `println()` gibi fonksiyonlar doğrudan çağrılabilir.
* Bazı fonksiyonlar değer alır ve/veya döner.
* Sonuç döndürmeyen fonksiyonlar bile aslında `Unit` tipi bir sonuç döner.
* Kotlin’de birçok hazır fonksiyon vardır, sadece çağırmak yeterlidir.
