### 🔢 **Sayı Tipleri**

#### **Tamsayılar (Integer Types)**

* **Byte** (8 bit): -128 → 127
* **Short** (16 bit): -32,768 → 32,767
* **Int** (32 bit): −2³¹ → 2³¹−1 *(en yaygın)*
* **Long** (64 bit): −2⁶³ → 2⁶³−1 *(en büyük tip)*

📝 Örnek:

```kotlin
val a = 1         // Int
val b = 2_000_000L // Long (L ile belirtilmiş)
val c: Byte = 15
```

#### **Ondalıklı Sayılar (Floating Point Types)**

* **Float** (32 bit): \~6-7 basamak hassasiyet
* **Double** (64 bit): \~14-16 basamak hassasiyet *(daha yaygın)*

📝 Örnek:

```kotlin
val pi = 3.14         // Double
val ratio: Float = 1.5f // Float (f ile belirtilmiş)
```

📏 Maksimum ve minimum değerler:

```kotlin
println(Int.MAX_VALUE)  // 2147483647
println(Int.SIZE_BYTES) // 4
```

---

### 🔡 **Karakter Tipi (Char)**

* Tek bir karakteri temsil eder (tek tırnak içinde yazılır).
* 2 byte (16 bit) boyutundadır.

📝 Örnek:

```kotlin
val letter = 'A'
val symbol = '$'
```

---

### ✅ **Mantıksal Tip (Boolean)**

* Sadece `true` veya `false` değerini alabilir.
* Koşullu ifadelerde kullanılır.

📝 Örnek:

```kotlin
val isOpen = true
```

---

### 🧵 **String Tipi**

* Çift tırnak içinde yazılan karakter dizileridir.

📝 Örnek:

```kotlin
val name = "Kotlin"
```

---

### 🎯 **Sonuç**

Kotlin'de en yaygın veri tipleri şunlardır:

* `Int`, `Long`, `Double`, `Float`, `Char`, `Boolean`, `String`
  `Long`, en büyük tamsayı türüdür. Türlerin **boyutlarını ve değer aralıklarını** bilmek, özellikle iş görüşmelerinde faydalıdır.
