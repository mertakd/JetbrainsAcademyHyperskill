### Fonksiyonlar Nedir? Ne İşe Yarar?

Hatırlarsan, daha önce fonksiyonlar konusuna değinmiştik. Peki ama **fonksiyon** tam olarak nedir ve nasıl kullanılır?

Fonksiyon, belirli bir işlemi gerçekleştiren bir komutlar dizisidir. Program içerisinde, bu fonksiyonun ismini yazarak onu çalıştırabiliriz. Fonksiyonlar, alt programlar gibi düşünülebilir; örneğin bir şeyi ekrana yazdırmak, bir sayının karekökünü almak gibi işleri yaparlar.

Aşağıdaki görsel, fonksiyonların nasıl çalıştığını basitçe gösteriyor. Temelde bu işleyiş, **girdi verisini alan ve bunu işleyerek bir çıktı üreten bir kutu** gibi düşünülebilir.

---

### Fonksiyonlara Girdi Vermek (Argümanlar)

Bir fonksiyonu kullanmak istediğimizde, **fonksiyonun ismini ve ardından parantezleri** yazarız. Eğer bu fonksiyon bir veya birden fazla girdi (argüman) alıyorsa, bu girdiler parantez içinde belirtilir.

Örneğin:

```kotlin
val text = "Merhaba"
println(text)
```

Burada `println` fonksiyonunu bir metinle birlikte çağırıyoruz. Bu fonksiyon aynı zamanda **hiçbir argüman almadan** da çalışabilir, bu durumda sadece bir alt satıra geçer:

```kotlin
println()
```

Genel kullanım biçimi şu şekildedir:

```kotlin
function1()           // Argüman almayan fonksiyon çağrısı
function2(arg1)        // Bir argüman alan fonksiyon
function3(arg1, arg2)  // İki argüman alan fonksiyon
```

---

### Fonksiyonların Sonuç Döndürmesi

Bazı fonksiyonlar, sadece işlem yapmakla kalmaz, aynı zamanda bir **sonuç (değer)** da döndürür. Bu değeri bir değişkene atayabiliriz:

```kotlin
val sonuc = function(arg)
```

Bu tarz fonksiyonlar, matematikteki fonksiyonlara benzer. Örneğin, bir sayının mutlak değerini döndüren `abs` fonksiyonu:

```kotlin
val sayi = -10
val pozitif = Math.abs(sayi)  // -10 veriyoruz, 10 döndürüyor
```

Bu örnekte `abs` fonksiyonu `Math` adlı bir grup içinde yer alıyor. Nokta (`.`) operatörünün sebebi bu: Kotlin’de bazı fonksiyonlar, bir gruba ait olarak tanımlanır ve bu yüzden `Math.abs()` şeklinde çağrılır. Şimdilik bu detaya çok takılma, ilerleyen konularda daha detaylı öğreneceğiz.

---

### Tüm Fonksiyonlar Bir Şey Döndürür mü?

Aslında, **evet**, her fonksiyon bir şey döndürür.

Örneğin:

```kotlin
val sonuc = println("merhaba")
println(sonuc)  // Çıktı: kotlin.Unit
```

Burada `println` fonksiyonu bir şey yazdırıyor ama görünürde bir değer döndürmüyor. Ancak Kotlin’de bu tür “boş” döndürmelerin karşılığı `Unit` adlı özel bir değerdir. `Unit`, başka programlama dillerinde gördüğümüz `void` türüne benzer. Yani fonksiyon bir şey döndürmüyor gibi görünse de teknik olarak `Unit` döndürür.
