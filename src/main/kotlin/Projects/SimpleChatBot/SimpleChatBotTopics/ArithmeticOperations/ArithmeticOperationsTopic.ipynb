### 🔢 **Kotlin'de Aritmetik Operatörler – Özet**

Kotlin, temel aritmetik işlemleri yapmak için 5 **binary (ikili)** operatör sunar:

#### ✅ **Temel Operatörler**

* **Toplama**: `+`
* **Çıkarma**: `-`
* **Çarpma**: `*`
* **Tam sayı bölme**: `/` → sadece tam kısmı verir (ondalık kısım atılır)
* **Mod (kalan)**: `%` → bölümden kalan değeri verir

#### 📌 Örnekler:

```kotlin
println(8 / 3)  // prints 2
println(10 % 3) // prints 1
```

---

### 🧮 **İfade Önceliği (Operator Precedence)**

İşlem sırası şu şekildedir (önceliği yüksekten düşüğe):

1. **Parantez** `()`
2. **Unary (+ / -)** → örnek: `-a`, `+b`
3. **Çarpma / Bölme / Mod** `* / %`
4. **Toplama / Çıkarma** `+ -`

**BODMAS** kuralı bu sırayı hatırlamaya yardımcı olur:
**Brackets, Orders, Division, Multiplication, Addition, Subtraction**

#### 📌 Örnek:

```kotlin
println(1 + 3 * 4 - 2) // prints 11 (çarpma önce yapılır)
println((1 + 3) * (4 - 2)) // prints 8 (parantez öncelikli)
```

---

### 🔢 **Basamak Ayırma / Sayı Çözümleme**

Sayının birler ve onlar basamağını almak için `%` ve `/` kullanılabilir:

```kotlin
println(54 % 10) // 4 (birler basamağı)
println((54 / 10) % 10) // 5 (onlar basamağı)
```

---

### ➕➖ **Unary Operatörler (Tekli)**

* `+` → sayıyı olduğu gibi bırakır
* `-` → sayıyı negatif yapar veya negatifse pozitife çevirir

```kotlin
println(+5)    // 5
println(-8)    // -8
println(-(100 + 4)) // -104
```

---

### 📌 **Sonuç**

* Kotlin'de aritmetik işlemler çok benzer şekilde yapılır.
* İşlem sırasına dikkat etmek önemlidir.
* Parantez kullanımı, işlemlerin önceliğini kontrol etmek için yararlıdır.
* Mod ve bölme işlemleri sayısal çözümlemelerde kullanılabilir.
