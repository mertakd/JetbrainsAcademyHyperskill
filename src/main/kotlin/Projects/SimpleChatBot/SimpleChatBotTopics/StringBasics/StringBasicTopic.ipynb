### 📚 **Kotlin'de String Türü – Özet**

#### 🔤 **String Nedir?**

* String, çift tırnak (`" "`) içinde sıfır veya daha fazla karakterden oluşur. Örneğin: `"John"`, `""`.
* Kotlin'de çok sık kullanılır (örneğin `"Hello World!"` gibi).

---

### 📏 **String Uzunluğu**

* `.length` özelliği ile string uzunluğu bulunur.

```kotlin
val text = "Kotlin"
println(text.length) // 6
```

---

### ➕ **String Birleştirme (Concatenation)**

* `+` operatörü ile iki string birleştirilir.

```kotlin
val fullName = "John" + " " + "Smith" // John Smith
```

* Sayılar ve diğer türlerle de birleştirme yapılabilir. Değerler otomatik olarak string’e dönüştürülür:

```kotlin
val code = "abc" + 10 + true // "abc10true"
```

⚠️ İlk operand bir `String` olmalı; aksi halde hata verir.

---

### 🔁 **String Tekrarı (repeat)**

* `.repeat(n)` ile bir string tekrar ettirilebilir.

```kotlin
println("Hello".repeat(3)) // HelloHelloHello
```

* Örnek: Haftalık program yazdırmak

```kotlin
println("Eat. Sleep. Code.\n".repeat(7))
```

---

### 🧾 **Raw String (Ham String)**

* Üç tırnak (`"""`) ile yazılır.
* Özel karakterlere gerek kalmaz, çok satırlı metinleri rahat yazmak için kullanılır.

```kotlin
val poem = """
    This is a raw string.
    No need to escape "quotes" or \n newlines.
""".trimIndent()
```

* `.trimIndent()` boşlukları ve ilk/son satırları temizler.

---

### 🧠 **Akılda Tutulması Gerekenler**

* `.length` ile uzunluk alınır.
* `+` ile birleştirme yapılır.
* `.repeat(n)` ile string tekrar ettirilir.
* `""" """` ile raw (ham) string yazılır.
* String içinde değişken kullanmak için string template: `"Hello, $name"`.
